# Salary Distribution Analysis (Week 1 Assignment)

This notebook analyzes a salary dataset using:
- Box plot (IQR method)
- Z-score
- Outlier detection


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dataset
salary = np.array([1000, 500, 800, 700, 700, 800, 700, 900, 750, 850])
salary_sorted = np.sort(salary)
salary_sorted
```

## Box Plot Statistics

```python
Q1 = np.percentile(salary, 25)
Q2 = np.percentile(salary, 50)
Q3 = np.percentile(salary, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

Q1, Q2, Q3, IQR, lower_bound, upper_bound
```

```python
print('Q1:', Q1)
print('Median(Q2):', Q2)
print('Q3:', Q3)
print('IQR:', IQR)
print('lower bound:', lower_bound)
print('upper bound:', upper_bound)
```

## Box Plot Visualization

```python
plt.boxplot(salary)
plt.title('Salary Distribution Box Plot')
plt.show()
```

## Z-Score Calculation

```python
mean = np.mean(salary)
std = np.std(salary)

z_scores = (salary - mean) / std

mean, std, z_scores
```

## Outlier Detection

```python
# Using IQR
outliers_iqr = salary[(salary < lower_bound) | (salary > upper_bound)]

# Using Z-score
outliers_z = salary[np.abs(z_scores) > 3]

outliers_iqr, outliers_z
```

## Conclusion
- No outliers detected
- Data is stable and well distributed
